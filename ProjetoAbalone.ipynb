# Sexo


Por conta dos valores não serem numéricos, foi necessário a troca para
correspondentes de 0 a 2, usado a imputação por vizinhança

```python
dfMacho = dfmexido[dfmexido["Sexo"]=="M"]
```

```python
dfFemea = dfmexido[dfmexido["Sexo"]=="F"]
```

```python
dfInfant = dfmexido[dfmexido["Sexo"]=="I"]
```

```python
LinhasMacho = dfMacho.index
dfmexido.loc[LinhasMacho,"Sexo"] = 0
```

```python
LinhasFemea = dfFemea.index
dfmexido.loc[LinhasFemea,"Sexo"] = 1
```

```python
LinhasInfant = dfInfant.index
dfmexido.loc[LinhasInfant,"Sexo"] = 2
```

```python
colunas_numericas = dfmexido.columns[df.dtypes != object].tolist()
coluna_a_imputar = 'Sexo'


imputer = KNNImputer(n_neighbors=10)
dfmexido[coluna_a_imputar] = imputer.fit_transform(dfmexido[colunas_numericas])


dfmexido[coluna_a_imputar] = np.round(dfmexido[coluna_a_imputar])


print(dfmexido)
```

```python
dfMacho = dfmexido[dfmexido["Sexo"]==0]
dfFemea = dfmexido[dfmexido["Sexo"]==1]
dfInfant = dfmexido[dfmexido["Sexo"]==2]
LinhasMacho = dfMacho.index
dfmexido.loc[LinhasMacho,"Sexo"] = "M"
LinhasFemea = dfFemea.index
dfmexido.loc[LinhasFemea,"Sexo"] = "F"
LinhasInfant = dfInfant.index
dfmexido.loc[LinhasInfant,"Sexo"] = "I"
```
